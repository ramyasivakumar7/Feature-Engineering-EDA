```python
col1=['member_id', 'loan_amnt','funded_amnt_inv','term','int_rate','grade','sub_grade','emp_length','annual_inc','verification_status','addr_state','dti','delinq_2yrs','inq_last_6mths','mths_since_last_delinq','open_acc','revol_bal','total_acc','total_rec_int','total_rec_late_fee','mths_since_last_major_derog','last_week_pay',
       'tot_coll_amt','acc_now_delinq','tot_cur_bal','total_rev_hi_lim','loan_status']
df2=df[col1]
df2
```

```python
df2.isnull().sum()
```

```python
df2['emp_length']=df2['emp_length'].str.replace("years|year"," ").str.replace("<","").str.replace(" ","").str.replace("+","")
df2['emp_length']=pd.to_numeric(df2['emp_length'],errors='coerce')

#downcast="integer" ---> if we give this, we will get dtype as int
#df2['emp_length'].value_counts(dropna=False)
```

```python
df2['emp_length'].fillna(0,inplace=True)
df2['emp_length'].isna().sum()
df2['emp_length']
```

```python
df2['term']=df2['term'].str.replace("months","")
df2['term']=pd.to_numeric(df2['term'],errors='coerce')    #-----> to convert string to numeric. errors=coerce is to ignore the errors
```

```python
df2['term']
```

```python
df2['last_week_pay']=df2['last_week_pay'].str.replace("th week|NA","")
df2['last_week_pay']=pd.to_numeric(df2['last_week_pay'],errors='coerce')
```

```python
df2['last_week_pay']
```

```python
df2['last_week_pay'].fillna(0,inplace=True)
df2['last_week_pay']
```

```python
df2.head()
```

```python
df2.isnull().any()
```

```python
df2['annual_inc'].fillna((df2['annual_inc'].mean()),inplace=True)
df2['tot_cur_bal'].fillna((df2['tot_cur_bal'].mean()),inplace=True)
df2['total_rev_hi_lim'].fillna((df2['total_rev_hi_lim'].mean()),inplace=True)
df2['tot_coll_amt'].fillna((df2['tot_coll_amt'].mean()),inplace=True)
```

```python
df2['mths_since_last_major_derog'].fillna((df2['mths_since_last_major_derog'].mode()[0]),inplace=True)    # generally mode[0] is given to replace al the values of nan to mode. we have to index in the series in mode
df2['acc_now_delinq'].fillna((df2['acc_now_delinq'].mode()[0]),inplace=True)
df2['delinq_2yrs'].fillna((df2['delinq_2yrs'].mode()[0]),inplace=True)
df2['inq_last_6mths'].fillna((df2['inq_last_6mths'].mode()[0]),inplace=True)
df2['mths_since_last_delinq'].fillna((df2['mths_since_last_delinq'].mode()[0]),inplace=True)
df2['open_acc'].fillna((df2['open_acc'].mode()[0]),inplace=True)
df2['total_acc'].fillna((df2['total_acc'].mode()[0]),inplace=True)
```
